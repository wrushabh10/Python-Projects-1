```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

olympics_df = pd.read_csv('/content/Olympics- visualization.csv')
olympics_df.columns
```

```python
condition = [(olympics_df['Medal'] > 1), (olympics_df['Medal'] == 0), (olympics_df['Medal'] == 1)]
choices = ['Positive', 'Negative','Neutral']

olympics_df['Sentiment'] = np.select(condition, choices, default='Negative')
olympics_df['Sentiment'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,6))
plt.pie(olympics_df['Sentiment'].value_counts(), labels = olympics_df['Sentiment'].value_counts().index, autopct = '%1.1f%%')
plt.show()
```

```python
conditions = [

     (olympics_df['Medal'] >1),

    (olympics_df['Medal'] == 0),

    (olympics_df['Medal']==1)]

choices = ['Positive', 'Negative','Neutral']

olympics_df['Sentiment'] = np.select(conditions,choices, default='Negative')

plt.figure(figsize=(9,9))
plt.pie(olympics_df["Sentiment"].value_counts(),labels=olympics_df["Sentiment"].value_counts().index,autopct='%1.1f%%', startangle = 90, shadow=True)
plt.legend(title ='Sentiment')

plt.show()
```
